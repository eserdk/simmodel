```python
def std(m=2 ** 32, a=1103515245):
    std.current = (a * std.current) % m
    yield std.current
std.current = 1
```

```python
import matplotlib.pyplot as plt
import random


def count(p):
    import math
    state_step = 0.01
    state_current = 0
    k = 0
    probability = []
    states = [0]
    parameters = []
    while k < 11:
        if p == 1:
            l = 1.75
        elif p == 2:
            l = 0.35 * k ** 2 + 9
        elif p == 3:
            l = 3.1 / (k + 7)
        Q = math.exp(-l * state_step) 
        state_current += state_step
        if Q < random.random():
            k += 1
            states.append(state_current)
        probability.append(1-Q)
        parameters.append(l*state_step)
    
    plt.step(states, range(12))
    plt.ylabel('States')
    plt.xlabel('Time')
    plt.show()
    
    plt.plot(probability)
    plt.ylabel('Probability')
    plt.xlabel('Time')
    plt.show()

    plt.plot(parameters)
    plt.ylabel('Parameters')
    plt.xlabel('Time')
    plt.show()
```
