```python
def std(m=2 ** 32, a=1103515245):
    std.current = (a * std.current) % m
    yield std.current
std.current = 1
```

```python
import random
from matplotlib import pyplot as plt

def count(p):
    import math
    time_step = 0.01
    time_current = 0
    times = [0]
    state = 0
    k = 1
    probability = []
    states = [0]
    parameters = []
    while k < 12:
        if p == 1:
            l = 5.5
            m = 0.8
        elif p == 2:
            l = 5.4 * k + 6
            m = 0.83 * k
        elif p == 3:
            l = 0.6 / pow(k + 1, 2)
            m = 3.1 / pow(k, 2)
        Q = math.exp(-(l+m)*time_step)
        time_current += time_step
        if Q < random.random():
            k += 1
            if l / (l + m) > random.random():
                state += 1
            else:
                state -= 1 if state >= 1 else 0
            times.append(time_current)
            states.append(state)
        probability.append(1-Q)
        parameters.append(l*time_step)
    
    plt.step(times, states)
    plt.ylabel('States')
    plt.xlabel('Time')
    plt.show()
    
    plt.plot(probability)
    plt.ylabel('Probability')
    plt.xlabel('Time')
    plt.show()

    plt.plot(parameters)
    plt.ylabel('Parameters')
    plt.xlabel('Time')
    plt.show()
```
